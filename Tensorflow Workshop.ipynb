```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

tf.reset_default_graph()
np.random.seed(101)
tf.set_random_seed(101)

X = tf.placeholder(tf.float32,name="X")
Y = tf.placeholder(tf.float32,name="Y")
myConstant = tf.constant(2,dtype=tf.float32,name="2")

random_data = 1000

expression = tf.add(tf.add(tf.multiply(tf.pow(X,myConstant),Y),Y),myConstant,name="expression")

writer = tf.summary.FileWriter('graphs/test1',tf.get_default_graph())
with tf.Session() as sess:
    # Iterating through all the epochs
    for epoch in range(random_data):
        x_random = np.random.rand(1,1)
        y_random = np.random.rand(1,1)
        print(sess.run(expression,feed_dict={X: x_random, Y:y_random}))
writer.close()        
```
